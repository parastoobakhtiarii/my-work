```python
from keras.layers import Dense, Dropout, Activation, LSTM, Convolution1D, MaxPooling1D, Flatten
from keras.models import Sequential
```

```python
data= data.values[:, 0:11]
print (data.shape)
```

```python
print (data[:11, :])
```

```python
y= data[0:, 0]
print (y.shape)
X= data[:3517,:]
print (X.shape)
y 
```

```python
X
```

```python
y= (y- min(y))/(max(y)- min(y))
print (y[3510:])
print (y[:7])
```

```python
print(np.min(X,axis=0))
```

```python
X= (X- np.min(X,axis=0))/(np.max(X, axis=0)-np.min(X,axis=0))
#X= (X- np.min(X))/(np.max(X)-np.min(X))
print (X.shape)
print (X[:3, :])
```

```python
time_steps= 50
X_new= np.zeros((X.shape[0] - time_steps +1, 100, X.shape[1]))
y_new= np.zeros((y.shape[0] -time_steps +1,))
for ix in range(X_new.shape[0]):
    for jx in range(time_steps):
        X_new[ix, jx, :]= X[ix +jx, :]
    y_new[ix]= y[ix + time_steps -1]
print (X_new.shape, y_new.shape)
```

```python
split = int(0.80*data.shape[0])
X_train = X_new[:split]
X_test = X_new[split:]

y_train = y_new[:split]
y_test = y_new[split:]

print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

```python
model = Sequential()
model.add(Convolution1D(64, 3, padding='same',input_shape= (100,11)))
model.add(MaxPooling1D(pool_size=2))
model.add(LSTM(100, return_sequences=True))
model.add(Dropout(0.2))
model.add(Convolution1D(32, 3, padding='same'))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(units=1))
model.add(Activation('linear'))

model.summary()
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
```

```python
history= model.fit(X_train, y_train, batch_size=128, epochs=100, validation_data = (X_test, y_test), verbose=2)
```

```python
score= model.evaluate(X_test, y_test, batch_size=128)
```

```python
print (score)
```

```python
y_pred= model.predict(X_test)
```

```python
plt.plot(y_pred, 'r-')
plt.plot(y_test, 'b-')
plt.show()
```

```python
import math
```

```python
MSE = np.square(np.subtract(y_test,y_pred)).mean() 
print (MSE)
```

```python
RMSE = math.sqrt(MSE)
print("Root Mean Square Error:\n")
print(RMSE)
```

```python
def mae(y_test, y_pred):
    y_test, y_pred = np.array(y_test), np.array(y_pred)
    return np.mean(np.abs(y_test - y_pred))
print(mae(y_test, y_pred))
```
